```python
from functools import total_ordering

@total_ordering
class Producto:
    def __init__(self, nombre: str, precio: float, stock: int):
        self.nombre = nombre
        self.precio = precio
        self.stock = stock
    def __str__(self):
        return f"{self.nombre} –{self.precio} - ({self.stock} unidades)"
    def __repr__(self):
        return f"Producto('{self.nombre}', {self.precio}, {self.stock})"
    def __eq__(self, otro):
        return (self.precio == otro.precio)
    def __lt__(self, otro):
        return (self.precio < otro.precio)
    def __add__(self, otro):
        return Producto(self.nombre, self.precio, self.stock + otro.stock)
    def __mul__(self, n):
        return Producto(self.nombre, self.precio * n, self.stock)


p1 = Producto('Arroz', 1200, 50) 
p2 = Producto('Arroz', 900, 30) 
p3 = Producto('Leche', 1200, 20) 
print(p1)              
print(repr(p2))        
print(p1 == p3)        
print(p2 < p1)         
print(sorted([p1, p2, p3]))
combinado = p1 + p2    
print(combinado)       
oferta = p1 * 0.8      
print(oferta)          

        

```

```python
class CuentaBancaria:
    def __init__(self, titular, saldo=0):
        self._titular = ""
        self._saldo = 0
        self.titular = titular
        self.saldo = saldo
    @property
    def titular(self):
        return self._titular
    @titular.setter
    def titular(self, valor):
        if len(valor) < 3:
            raise ValueError("Nombre muy corto")
        self._titular = valor
    @property
    def saldo(self):
        return self._saldo
    @saldo.setter
    def saldo(self, valor):
        if valor < 0:
            raise ValueError("Saldo no puede ser negativo")
        self._saldo = valor
    def depositar(self, monto):
        if monto > 0:
            self._saldo = self._saldo + monto
        else:
            raise ValueError("Monto inválido")
    def retirar(self, monto):
        if monto > 0 and self._saldo - monto >= 0:
            self._saldo = self._saldo - monto
        else:
            raise ValueError("No se puede retirar")
    def __str__(self):
        return "Cuenta de " + self._titular + " | Saldo: $" + str(self._saldo)
    def __repr__(self):
        return "CuentaBancaria(titular='" + self._titular + "', saldo=" + str(self._saldo) + ")"
    @property
    def aprobada(self):
        return self._saldo >= 10000
cuenta = CuentaBancaria('Ana López', 50000) 
print(cuenta)              # Cuenta de Ana López | Saldo: $50000  
cuenta.depositar(20000) 
print(cuenta.saldo)        # 70000 
cuenta.retirar(15000) 
print(cuenta)              # Cuenta de Ana López | Saldo: $55000 
# Validaciones 
try: 
    cuenta.saldo = -1000   # debe lanzar ValueError 
except ValueError as e: 
    print(e)               # Saldo no puede ser negativo 
 
try: 
    CuentaBancaria('AB', 0) # titular muy corto 
except ValueError as e: 
    print(e) 
 
print(cuenta.aprobada)     # True 
```

```python
import math

class Figura:
    def __init__(self, color):
        self.color = color

    def area(self):
        raise NotImplementedError()

    def perimetro(self):
        raise NotImplementedError()

    def __str__(self):
        return "Figura " + self.color + " | Área: " + str(round(self.area(), 2)) + " | Perímetro: " + str(round(self.perimetro(), 2))

    def __repr__(self):
        return "Figura(color='" + self.color + "')"

    # BONUS: comparar por área
    def __lt__(self, other):
        return self.area() < other.area()


class Circulo(Figura):
    def __init__(self, color, radio):
        super().__init__(color)
        if radio <= 0:
            raise ValueError("Error")
        self.radio = radio

    def area(self):
        return math.pi * self.radio * self.radio

    def perimetro(self):
        return 2 * math.pi * self.radio


class Rectangulo(Figura):
    def __init__(self, color, ancho, alto):
        super().__init__(color)
        if ancho <= 0 or alto <= 0:
            raise ValueError("Error")
        self.ancho = ancho
        self.alto = alto

    def area(self):
        return self.ancho * self.alto

    def perimetro(self):
        return 2 * (self.ancho + self.alto)

    @property
    def es_cuadrado(self):
        return self.ancho == self.alto


# ---- PRUEBAS ----
c = Circulo('rojo', 5)
r = Rectangulo('azul', 4, 6)
cuad = Rectangulo('verde', 5, 5)

print(c)
print(r)

print(cuad.es_cuadrado)   # True
print(r.es_cuadrado)      # False

# Polimorfismo
figuras = [c, r, cuad]
for f in figuras:
    print('Área:', round(f.area(), 2))

# Ordenar por área
figuras.sort()
print("Ordenadas por área:")
for f in figuras:
    print(f)
```

```python
class Producto:
    def __init__(self, nombre, precio, stock):
        self.nombre = nombre
        self.precio = precio
        self.stock = stock
    def __str__(self):
        return self.nombre + " – $" + str(self.precio) + " (" + str(self.stock) + " unidades)"
    def __eq__(self, other):
        return self.nombre == other.nombre and self.precio == other.precio and self.stock == other.stock
class Inventario:
    def __init__(self):
        self.productos = []
    def agregar(self, producto):
        self.productos.append(producto)
    def __len__(self):
        return len(self.productos)
    def __getitem__(self, i):
        return self.productos[i]
    def __contains__(self, producto):
        return producto in self.productos
    def __iter__(self):
        return iter(self.productos)
    def __str__(self):
        texto = ""
        for p in self.productos:
            texto = texto + str(p) + "\n"
        return texto
    def __repr__(self):
        return "Inventario(n_productos=" + str(len(self.productos)) + ")"
    @property
    def total_valor(self):
        total = 0
        for p in self.productos:
            total = total + p.precio * p.stock
        return total
    @property
    def mas_caro(self):
        return max(self.productos, key=lambda p: p.precio)
    def buscar(self, nombre):
        resultado = []
        for p in self.productos:
            if nombre.lower() in p.nombre.lower():
                resultado.append(p)
        return resultado
inv = Inventario() 
inv.agregar(Producto('Arroz', 1200, 50)) 
inv.agregar(Producto('Leche', 950, 30)) 
inv.agregar(Producto('Pan', 1800, 100)) 
inv.agregar(Producto('Panetón', 3500, 10)) 
print(len(inv))            
# 4 
print(inv[0])              
# Arroz – $1200 (50 unidades) 
p_buscar = Producto('Leche', 950, 30) 
print(p_buscar in inv)     
# True 
for p in inv:
    print(p)

print(inv.total_valor)     
print(inv.mas_caro)        

resultados = inv.buscar('pan') 
for r in resultados:
    print(r)
```
